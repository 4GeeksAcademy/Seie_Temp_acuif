```python
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
import pandas as pd
def test_stationary(timeseries, plot=True, window=12):
    """
    Aplica la prueba de Dickey-Fuller a una serie temporal y muestra resultados.
    Opcionalmente grafica la serie con su media mvil.
    """
    
    # Validacin
    if not isinstance(timeseries, pd.Series):
        raise TypeError("Se esperaba una Serie de Pandas. Usa total_data['Depth_to_Groundwater'], no el DataFrame completo.")
    print(" Resultados de la prueba de Dickey-Fuller:\n")
    
    # Prueba de Dickey-Fuller
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=["Estadstico de prueba", "p-valor", "# lags usados", "# observaciones"])
    
    # Valores crticos
    for key, value in dftest[4].items():
        dfoutput[f"Valor crtico ({key})"] = value
    print(dfoutput)
    # Grfico de la serie y su media mvil
    if plot:
        rolmean = timeseries.rolling(window).mean()
        rolstd = timeseries.rolling(window).std()
        plt.figure(figsize=(10, 5))
        plt.plot(timeseries, label="Serie original", color='blue')
        plt.plot(rolmean, label=f"Media mvil ({window})", color='red')
        plt.plot(rolstd, label=f"Desvo mvil ({window})", color='green', linestyle='--')
        plt.legend()
        plt.title("Evolucin de la serie y media mvil")
        plt.show()
# Asegrate de tener el ndice en datetime
total_data["Date"] = pd.to_datetime(ts)
total_data.set_index("Date", inplace=True)
# Ejecuta la prueba sobre la serie de ventas
test_stationary(total_data["Depth_to_Groundwater"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(ts, model='adittive', period=12)

trend = decomposition.trend

fig, ax = plt.subplots(figsize=(10, 5))
sns.lineplot(data=ts, label='Original', ax=ax)
sns.lineplot(data=trend, label='Trend', ax=ax)

# Improve layout and show the plot
plt.legend()
plt.title('Seasonal Decomposition of Time Series')
plt.xlabel('Time')
plt.ylabel('Values')
plt.tight_layout()
plt.show()
```

```python
seasonal = decomposition.seasonal

fig, axis = plt.subplots(figsize = (10, 5))

sns.lineplot(data = ts)
sns.lineplot(data = seasonal)

plt.tight_layout()

plt.show()
```

```python
residual = decomposition.resid

fig, axis = plt.subplots(figsize = (10, 5))

sns.lineplot(data = ts)
sns.lineplot(data = residual)

plt.tight_layout()

plt.show()
```
